## Pré processamento

* Funções para fazer o tratamento

```python
# retirar url's
url_re = re.compile('https?://(www.)?\w+\.\w+(/\w+)*/?')
```

```python
# pontuacao
punc_re = re.compile('[%s]' % re.escape(string.punctuation))
```

```python
# numeros
num_re = re.compile('(\\d+)')
```

```python
# alfa numericos
alpha_num_re = re.compile("^[a-z0-9_.]+$")
```

```python
# mencoes
mention_re = re.compile('@(\w+)')
```

## Exemplos de tratamento

```python
# deixar minusculo
texto_ajustado = df['texto_do_twitter'].apply(lambda x: x.lower())
texto_ajustado
```

```python
# limpar urls
texto_ajustado[texto_ajustado.str.contains('http')]
```

```python
texto_ajustado[texto_ajustado.str.contains('http')].apply(lambda x: url_re.sub(' ', x))
```

```python
texto_ajustado = texto_ajustado.apply(lambda x: url_re.sub(' ', x))
```

```python
# Ajustar menções com @
texto_ajustado = texto_ajustado.apply(lambda x: mention_re.sub(' ', x))
texto_ajustado
```

```python
# ajustar pontuação
texto_ajustado = texto_ajustado.apply(lambda x: punc_re.sub(' ', x))
texto_ajustado
```

```python
# limpar numeração
texto_ajustado = texto_ajustado.apply(lambda x: num_re.sub(' ', x))
texto_ajustado
```

```python
# ajustar acentuação
texto_ajustado = texto_ajustado.apply(lambda x: normalize('NFKD', x).encode('ASCII', 'ignore').decode('ASCII'))
texto_ajustado
```

```python
# limpar textos alfanumericos
def preprocessamento_texto(texto, n):
  '''Ajustar caracteres alfa numéricos e palavras menores que n
     texto: texto a ser ajustado
     n: tamanho mínimo das palavras'''
  list_pos = 0
  cleaned_str = ''
  for word in texto.split():
      if list_pos == 0:
          if alpha_num_re.match(word) and len(word) > n:
              cleaned_str = word
          else:
              cleaned_str = ' '
      else:
          if alpha_num_re.match(word) and len(word) > n:
              cleaned_str = cleaned_str + ' ' + word
          else:
              cleaned_str += ' '
      list_pos += 1
  return cleaned_str
```

```python
texto_ajustado = texto_ajustado.apply(lambda x: preprocessamento_texto(x, 3))
texto_ajustado
```

```python
# ajustar palavras mto usadas
def remove_stops(texto):
    # expects a string
    stops = set(STOP_WORDS)
    list_pos = 0
    cleaned_str = ''
    text = texto.split()
    for word in text:
        if word not in stops:
            # rebuild cleaned_str
            if list_pos == 0:
                cleaned_str = word
            else:
                cleaned_str = cleaned_str + ' ' + word
            list_pos += 1
    return cleaned_str
```

```python
list(STOP_WORDS)[0:15]
```

```python
texto_ajustado = texto_ajustado.apply(lambda x: remove_stops(x))
texto_ajustado
```

```python
df['texto_twitter_ajustado'] = texto_ajustado
```
