# Sistema de recomendação de roupas


```python
import pandas as pd
import numpy as np
from itertools import product
from math import factorial
```

# Peças de roupas

```python
n_cabeca = 5
n_busto = 10
n_pernas = 8
n_pes = 7
```

```python
combinacoes_possiveis = n_busto * n_cabeca * n_pes * n_pernas
combinacoes_possiveis
```

# Criar os ids das roupas


```python
id_cabeca = ['cabeca'+str(i+1) for i in range(n_cabeca)]
id_busto = ['busto'+str(i+1) for i in range(n_busto)]
id_pernas = ['pernas'+str(i+1) for i in range(n_pernas)]
id_pes = ['pes'+str(i+1) for i in range(n_pes)]
id_busto
```

```python
lista_roupas = [id_cabeca, id_busto, id_pernas, id_pes]
```

```python
combinacoes_roupas = list(product(*lista_roupas))
```

```python
df_combinacaoes_roupas = pd.DataFrame(combinacoes_roupas, columns=['cabeca_sugestao', 'busto_sugestao',
                                               'pernas_sugestao', 'pes_sugestao'])
```

```python
df_combinacaoes_roupas
```

```python
df_combinacoes_roupas = df_combinacaoes_roupas \
                                .reset_index() \
                                .rename(columns = {'index': 'id_combinacao'})
df_combinacoes_roupas
```

```python
df_combinacoes_roupas.head()
```

```python
n_amostras = 10
amostras_roupas = df_combinacoes_roupas.sample(n_amostras).reset_index(drop=True)
amostras_roupas
```
